```python
data = pd.read_csv('telco_customer_churn.csv') # Memuat data dari file CSV 'telco_customer_churn.csv' ke dalam DataFrame

data.head() # Menampilkan lima baris pertama dari DataFrame
```

a. Gambarlah histogram dari kolom 'Tenure', dan pisahkan dengan warna berdasarkan Churn. Namun, pada visualisasi tersebut, tambahkan pula 'Density Plot' di atas Histogram.

```python
fig = plt.figure() # Membuat objek figure (fig) untuk plot
ax = sns.histplot(data = df, x='tenure') # Membuat histogram dengan seaborn (sns) untuk kolom 'tenure' dalam DataFrame 'df'
plt.show() # Menampilkan plot histogram
```

```python
# Visualisasi Histogram Tenure dengan Density Plot (Dipisahkan berdasarkan Churn)
plt.figure(figsize=(12,5))
plt.figure(facecolor='white')
sns.set(style='whitegrid')
sns.histplot(data=data, x='tenure', hue='Churn', kde=True)
plt.show()
```

b. Buatlah Stacked Bar Chart yang memperlihatkan rasio dari kolom 'InternetService' terhadap customer-customer yang Churn dan Non Churn. 

```python
# Persentase Churn berdasarkan Jenis Layanan Internet
pd.crosstab(df['InternetService'], df['Churn']).apply(lambda x: x*100/x.sum(), axis = 1)
```

```python
pd.crosstab(df['InternetService'], df['Churn']).apply(lambda x: x*100/x.sum(), 
axis = 1).plot(kind = 'bar', stacked = True)
plt.show()
```

c. Jawablah pertanyaan ini: bagi customer yang Churn, kebanyakan dari mereka memiliki InternetService tipe apa?

```python
# Churn berdasarkan Jenis Layanan Internetplt.figure(figsize=(8, 6))
sns.set(style='whitegrid')
ax=sns.countplot(data=data, x='InternetService', hue='Churn', saturation=1, alpha=0.9)
ax.set_title('Churn by Internet Service')
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), ha='center', va='top', color='white', size=10)
```

a. Buatlah visualisasi scatterplot menggunakan Sub Plots (materi P3 Part 2). Terdapat 2 Sub Plots yang akan dibuat, dengan ketentuan berikut:

Scatterplot pertama: sumbu-x: body_mass_g, sumbu-y: flipper_length_mm

Scatterplot kedua: sumbu-x: body_mass_g, sumbu-y: bill_length_mm

Scatterplot diberi warna berdasarkan kolom island. Terserah menggunakan color palette apa pun.

```python
# Hubungan Antara Berat Tubuh dan Panjang Flipper serta Panjang Paruh pada Berbagai Pulau
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
sns.scatterplot(data=df3, x='body_mass_g', y='flipper_length_mm', hue='island')
plt.title('Berat Tubuh vs Panjang Flipper')
plt.xlabel('Berat Tubuh (g)')
plt.ylabel('Panjang Flipper (mm)')

plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
sns.scatterplot(data=df3, x='body_mass_g', y='bill_length_mm', hue='island')
plt.title('Berat Tubuh vs Panjang Paruh')
plt.xlabel('Berat Tubuh (g)')
plt.ylabel('Panjang Paruh (mm)')

plt.show()
```
